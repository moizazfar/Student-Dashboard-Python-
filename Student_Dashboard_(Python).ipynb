# **STUDENT DASHBOARD**

```python
def add_student(students:list):
  name = input("Enter student name to add student to list: ").title().strip()
  students.append(name)
  print("*STUDENT ADDED SUCCESSFULLY*")
```

```python
def delete_student(students:list):
  name = input("Enter student name to delete student from list: ").title().strip()
  students.remove(name)
  print("*STUDENT REMOVED SUCCESSFULLY*")
```

```python
def add_student_marks(students:list,student_marks:list):
  mark_list=[]
  name = input("Enter name of student to add their marks: ").title().strip()
  if name in students:
    for i in range(1,4):
      marks = int(input(f"Enter marks of subject {i}: "))
      mark_list.append(marks)
    student_marks.append(mark_list)
    print("Marks added successfully")
  else:
    print("Student not present")
```

```python
def view_student_details(students:list,student_marks:list):
  print('''______________________________________

      ALL STUDENTS WITH THEIR MARKS
______________________________________
              ''')
  print("\nStudent details...")
  for i, name in enumerate(students):
    if i < len(student_marks) and student_marks[i]:
      total=sum(student_marks[i])
      per=total/300*100
      print(f"{i+1}) Name: {name} | Marks: {student_marks[i]} | Total Marks: {total} | Percentage: {round(per,2)}")
    else:
      print(f"{i+1}) Name: {name} | Marks: Not added yet")
```

```python
def student_dashboard(students:list,student_marks:list):
  while True:
    print('''
    ______________________________________

              Student Dashboard
    ______________________________________

      PRESS 1 TO ADD STUDENT
      PRESS 2 TO DELETE STUDENT
      PRESS 3 TO ADD STUDENT MARKS
      PRESS 4 TO VIEW STUDENT
      PRESS 5 TO STOP

    ______________________________________
      ''')
    press = int(input("Press number: "))
    if press==1:
      add_student(students)
    elif press==2:
      delete_student(students)
    elif press==3:
      add_student_marks(students,student_marks)
    elif press==4:
      view_student_details(students,student_marks)
    elif press==5:
      break
    else:
      print("Invalid Press")
```

```python
student=[]
student_mark=[]
student_dashboard(student,student_mark)
```
